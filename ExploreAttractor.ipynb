```python
def plotx(df):
    plt.plot(df['t'], df['x'])
    
def ploty(df):
    plt.plot(df['t'], df['y'])
    
def plotz(df):
    plt.plot(df['t'], df['z'])

def plotxy(df):
    plt.plot(df['x'], df['y'])
def plotyz(df):
    plt.plot(df['y'], df['z'])
def plotxz(df):
    plt.plot(df['x'], df['z'])

def plot3d(df):
    fig = plt.figure()
    ax = fig.gca(projection='3d')
    ax.plot_trisurf(df['x'], df['y'], df['z'])
    
    
```

```python
attract.points=100
df = attract.evolve()
plot3d(df)
```

Fewer Points lowers the accuracy

```python
attract.points=10000
df = attract.evolve(np.array([0.1,0.,0.]),1)
plot3d(df)
```

Lower Order Integration Lowwers acccuracy

```python
df = attract.evolve(np.array([1.1,2.,4.]),4)
plot3d(df)
```

```python
df = attract.evolve(np.array([-1.1,2.,400]),4)
plot3d(df)
```
